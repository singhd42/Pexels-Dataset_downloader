```python
# Step 2: Function to download images of single faces from Pexels
def download_images(query="single face", num_images=20000, output_dir="pexels_faces_images"):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    total_downloaded = 0
    page = 1
    
    while total_downloaded < num_images:
        params = {
            "query": query,
            "per_page": 80,  # Maximum images per request
            "page": page
        }
        
        response = requests.get("https://api.pexels.com/v1/search", headers=headers, params=params)
        data = response.json()
        
        if 'photos' not in data or not data['photos']:
            print("No more images found.")
            break
        
        for index, photo in enumerate(data['photos']):
            if total_downloaded >= num_images:
                break
            
            image_url = photo['src']['original']
            image_data = requests.get(image_url).content
            image_filename = os.path.join(output_dir, f"face_image_{total_downloaded + 1}.jpg")
            
            with open(image_filename, 'wb') as image_file:
                image_file.write(image_data)
            
            total_downloaded += 1
            print(f"Downloaded {image_filename}")
        
        page += 1
    
    print(f"Total images downloaded: {total_downloaded}")
```

```python
# Step 3: Function to download videos of single faces from Pexels
def download_videos(query="single face", num_videos=20000, output_dir="pexels_faces_videos"):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    total_downloaded = 0
    page = 1
    
    while total_downloaded < num_videos:
        params = {
            "query": query,
            "per_page": 80,  # Maximum videos per request
            "page": page
        }
        
        response = requests.get("https://api.pexels.com/videos/search", headers=headers, params=params)
        data = response.json()
        
        if 'videos' not in data or not data['videos']:
            print("No more videos found.")
            break
        
        for index, video in enumerate(data['videos']):
            if total_downloaded >= num_videos:
                break
            
            video_url = video['video_files'][0]['link']  # Download the first version of the video
            video_data = requests.get(video_url).content
            video_filename = os.path.join(output_dir, f"face_video_{total_downloaded + 1}.mp4")
            
            with open(video_filename, 'wb') as video_file:
                video_file.write(video_data)
            
            total_downloaded += 1
            print(f"Downloaded {video_filename}")
        
        page += 1
    
    print(f"Total videos downloaded: {total_downloaded}")
```
